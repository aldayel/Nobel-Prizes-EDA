# Age and Gender Related Visualizations 

```python
# Slicing only the year out of Birthday and excluding month and day

df['Birth_Year'] = df['Birthday'].str[0:4]
```

```python
# Replacing Missing values by 0

df['Birth_Year'] = df['Birth_Year'].replace(to_replace="nan", value=0)
```

```python
# Convert to int64/float64

df['Birth_Year'] = df['Birth_Year'].apply(pd.to_numeric)
```

```python
# Age Prize awarded is (year awarded - year born)

df['Age'] = df['Year'] - df['Birth_Year']
```

## Data Exploration by Country

```python
# generate country code  based on country name 
import pycountry 

def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
           # .alpha_3 means 3-letter country code 
           # .alpha_2 means 2-letter country code
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
# create a column for code 
df['Code']=alpha3code(df.Country)
df.head()
```
